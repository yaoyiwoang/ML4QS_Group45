```python
def aggregate_data_multiple_columns(df, timestamp_column, interval, features, trial_num, experimenter):

    # Create a copy of the DataFrame to avoid modifying the original data
    df_copy = df.copy()

    # Create a column that indicates the time interval group
    df_copy['interval_group'] = (df_copy[timestamp_column] // interval) * interval
    
    # Define the aggregation dictionary
    aggregation_functions = {col: ['max', 'min', 'median', 'mean', 'std'] for col in features}
    
    # Group by the interval group and aggregate the data
    aggregated = df_copy.groupby('interval_group').agg(aggregation_functions)
    
    # Flatten the MultiIndex columns
    aggregated.columns = ['_'.join(col).strip() for col in aggregated.columns.values]
    
    # Reset the index to turn the interval groups back into a column
    aggregated.reset_index(inplace=True)
    
    # Rename the 'interval_group' column to 'time_interval'
    aggregated.rename(columns={'interval_group': 'timestamp'}, inplace=True)
    
    # Handle the last time period to include it even if it's not a full interval
    max_time = df[timestamp_column].max()
    last_group = (max_time // interval) * interval
    if max_time > last_group:
        last_group_data = df_copy[df_copy[timestamp_column] > last_group]
        if not last_group_data.empty:
            last_group_stats = last_group_data[features].agg(aggregation_functions).T
            new_columns = {}
            for index, row in last_group_stats.iterrows():
                variable_name = index
                for col in last_group_stats.columns:
                    new_columns[col] = f"{variable_name}_{col}"  # Construct new column name
            last_group_stats.rename(columns=new_columns, inplace=True)
            last_group_stats.drop_duplicates(inplace=True)
            last_group_stats['timestamp'] = max_time
            aggregated = pd.merge(aggregated, last_group_stats, how='outer')
            aggregated['trial'] = trial_num
            aggregated['experimenter'] = experimenter
    return aggregated
```

```python
def merge_pd_horizontally(pd_list, key_cols=['timestamp', 'trial', 'experimenter']):
    
    merged_df = pd.concat(pd_list, axis=1, join='inner')
    merged_df = merged_df.loc[:,~merged_df.columns.duplicated()]
    new_columns = key_cols + [col for col in merged_df.columns if col not in key_cols]
    merged_df = merged_df[new_columns]
    
    return merged_df
```

```python
merge_pd = merge_pd_horizontally([yaoyi_trial_1_brigtness_agg, yaoyi_trial_1_audio_spectrum_agg])
```

```python
def plot_combined_boxplots(df, columns_to_plot, groupby_cols=['trial', 'experimenter']):
    """
    Plot combined box plots for specified columns grouped by 'trial' and 'experimenter'.
    
    Args:
        df (pd.DataFrame): DataFrame containing the data.
        columns_to_plot (list): List of columns to plot.
        groupby_cols (list): List of columns to group by.
    """
    # Create a new column combining 'trial' and 'experimenter'
    df['group'] = df[groupby_cols].astype(str).agg('_'.join, axis=1)
    
    # Initialize the figure
    plt.figure(figsize=(14, 10))
    
    # Iterate over columns to plot
    melted_df_list = []
    for col in columns_to_plot:
        # Melt the DataFrame to have a long format for each column to plot
        melted_df = df.melt(id_vars=['group'], value_vars=[col], 
                            var_name='Measurement', value_name='Value')
        # Create a new column for combined label
        melted_df['combined'] = melted_df['group'] + '_' + col
        melted_df_list.append(melted_df)
    
    # Concatenate all melted DataFrames
    final_melted_df = pd.concat(melted_df_list)
    
    # Create box plot
    sns.boxplot(x='combined', y='Value', data=final_melted_df)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Add labels and title
    plt.xlabel('Group_Measurement')
    plt.ylabel('Value')
    plt.title('Combined Box plots by Group and Measurement')
    
    # Show plot
    plt.show()
```

```python
plot_combined_boxplots(merge_pd, ['BrightnessEV_mean', 'Peak-Frequency_mean'])
```
